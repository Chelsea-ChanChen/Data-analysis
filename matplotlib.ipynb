```python
import matplotlib
```

```python
x=list(range(2,26,2))
y=[15,13,14.5,17,20,25,26,24,22,18,15,24]
```

```python
fig=plt.figure(figsize=(10,5),dpi=200) #10表示横轴大小，5表示纵轴大小。dpi表示每英寸上像素点的个数
plt.xticks(range(2,26,2)) #设置x轴的刻度
plt.tick_params(axis='x', rotation=45, labelsize=10, color='red') # 旋转、调整大小和颜色
# xtick_lables=[i/2 for i in range(4,49)]
# plt.xticks(range(25,50))
plt.plot(x,y) #传入x和y，通过plot绘制出折线图
plt.savefig("./tl.png") # 保存图像要放在show的前面
plt.show() #在执行程序的时候展示图形
```

```python
plt.yticks(range(min(y),max(y)+1))
plt.plot(x,y) 
plt.show()
```
